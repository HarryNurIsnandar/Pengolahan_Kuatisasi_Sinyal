```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
A = 1 # Amplitudo
Fa = 200 # Hz (Frekuensi Analog)
Fs = 3000 # Hz (Frekuensi Sampling)
N = int(5 * Fs/Fa) # Jumlah Sampel
tstep = 1 / Fs #Sample Interval Waktu
fstep = Fs / N #Sample Interval Frekuensi
t = np.linspace(0, (N-1) * tstep, N) # Rentang Nilai Waktu
f = np.linspace(0, (N-1) * fstep, N) # Rentang Nilai Frekuensi
```

```python
sn1 = A * np.sin (2 * np.pi * Fa * t)
sn2 = (4 * A) * np.sin(2 * np.pi * (3 * Fa) * t)
snt = sn1 + sn2
```

```python
plt.plot(t, sn1)
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.title("Sinyal 1")
```

```python
plt.plot(t, sn2)
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.title("Sinyal 2")
```

```python
plt.plot(t, snt)
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.title("Sinyal Tergabung")
```

```python
save_path = 'D:\\kuliah\\smt 7\\Pengolahan Sinyal Digital\\UTS_SAMPLING\\Pengolahan_Kuatisasi_Sinyal'
os.makedirs(save_path, exist_ok=True)
np.savez(os.path.join(save_path, 'sinyal_gabung.npz'), t=t, sn1=sn1, sn2=sn2, snt=snt)
```

```python
file_path = os.path.join(save_path, 'sinyal_gabung.npz')
data = np.load(file_path)
sng = data['snt']
t = data['t']
```

```python
plt.plot(t, sng, color="green")
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.title("Sinyal Tergabung Hasil")
```

```python
z = np.exp(-1j * 2 * np.pi * np.outer(f, t))
z_transform = np.dot(sng, z)
z_mag = np.abs(z_transform)
f_plot = f[:N//2]
z_mag_plot = z_mag[:N//2]
```

```python
plt.plot(f_plot, z_mag_plot, color="red")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Amplitudo")
plt.title("Transformasi Z")
```
